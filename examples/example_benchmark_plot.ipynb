```python

from crystaldiskmark_parser.parser import parse

import matplotlib.pyplot as plt
```

```python
result_wd = parse("../test/data/CrystalDiskMark_20210622162528 WD Blue 3D 1TB WDS100T2B0A.txt")
result_samsung = parse("../test/data/CrystalDiskMark_20210622163451 SAMSUNG 840 EVO 120GB.txt")

data = [result_wd.read_results[0].rate, result_samsung.read_results[0].rate]
con = plt.bar(["WD Blue 3D 1TB", "Samsung 840 EVO 120GB"], data)
plt.bar_label(con)
plt.title("Read Rate of SSDs")
plt.ylabel(f"Rate [{result_wd.read_results[0].unit_rate}]")

plt.show()
```
